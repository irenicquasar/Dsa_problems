# max w1

```python
def find_max_w1_triplet(A, B, C):
    max_diff = max(
        abs(A[-1] - B[0]),
        abs(A[-1] - C[0]),
        abs(B[-1] - A[0]),
        abs(B[-1] - C[0]),
        abs(C[-1] - A[0]),
        abs(C[-1] - B[0])
    )
    
    if max_diff == abs(A[-1] - B[0]):
        return len(A) - 1, 0, 0
    elif max_diff == abs(A[-1] - C[0]):
        return len(A) - 1, 0, 0
    elif max_diff == abs(B[-1] - A[0]):
        return 0, len(B) - 1, 0
    elif max_diff == abs(B[-1] - C[0]):
        return 0, 0, len(C) - 1
    elif max_diff == abs(C[-1] - A[0]):
        return 0, len(B) - 1, 0
    else:
        return 0, 0, len(C) - 1
    
# Example usage
A = [1, 4, 7, 10]
B = [20,23,26,299]
C = [3, 6, 9, 12]

i, j, k = find_max_w1_triplet(A, B, C)
print("Indices i, j, k:", i, j, k)
print("Maximum w1:", max(abs(A[i] - B[j]), abs(A[i] - C[k]), abs(B[j] - C[k])))

```

# min w1

```python
#min w1
def find_min_w1_triplet(A, B, C):
    i, j, k = 0, 0, 0
    max_diff = float('inf')

    while i < len(A) and j < len(B) and k < len(C):
        diff1 = abs(A[i] - B[j])
        diff2 = abs(B[j] - C[k])
        diff3 = abs(C[k] - A[i])

        current_diff = max(diff1, diff2, diff3)
        if current_diff < max_diff:
            max_diff = current_diff
            i_min, j_min, k_min = i, j, k

        # Trying to reduce the maximum of the three differences
        if diff1>=diff2 and diff1>=diff3:
            if A[i] <= B[j]:
                i += 1
            else:
                j += 1
        elif diff2>=diff1 and diff2>=diff3:
            if B[j] <= C[k]:
                j += 1
            else:
                k += 1
        else:  # diff3>=diff1 and diff3>=diff2
            if C[k] <= A[i]:
                k += 1
            else:
                i += 1

    return i_min, j_min, k_min
# Example usage
A = [1, 4, 7, 10]
B = [20,23,26,299]
C = [3, 6, 9, 12]

i, j, k = find_min_w1_triplet(A, B, C)
print("Indices i, j, k:", i, j, k)
print("Minimum w1:", min(abs(A[i] - B[j]), abs(A[i] - C[k]), abs(B[j] - C[k])))
```

# min w2

```python
#min w2
def find_min_w2_triplet(A, B, C):
    i, j, k = 0, 0, 0
    min_diff = float('inf')

    while i < len(A) and j < len(B) and k < len(C):
        diff1 = abs(A[i] - B[j])
        diff2 = abs(B[j] - C[k])
        diff3 = abs(C[k] - A[i])

        current_diff = min(diff1, diff2, diff3)
        # this will give me index of the minimum difference
        if current_diff < min_diff:
            min_diff = current_diff
            i_min, j_min, k_min = i, j, k

        # Trying to reduce the maximum of the three differences
        if diff1 < diff2 and diff1 < diff3:
            if A[i] < B[j]:
                i += 1
            elif A[i] == B[j]:
                return i, j, k
            else:
                j += 1
        elif diff2 < diff1 and diff2 < diff3:
            if B[j] < C[k]:
                j += 1
            elif B[j] == C[k]:
                return i, j, k
            else:
                k += 1
        else:  # diff3 < diff1 and diff3 < diff2
            if C[k] < A[i]:
                k += 1
            elif A[i] == C[k]:
                return i, j, k
            else:
                i += 1
    return i_min, j_min, k_min

# Example usage
A = [2, 5, 17, 50]
B = [12,15, 17, 38]
C = [7, 23, 47, 72]

i, j, k = find_min_w2_triplet(A, B, C)
print("Indices i, j, k:", i, j, k)
print("Minimum w2:", min(abs(A[i] - B[j]), abs(A[i] - C[k]), abs(B[j] - C[k])))
```
